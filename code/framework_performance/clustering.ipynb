```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
from matplotlib import rc
from matplotlib import gridspec

rc('text', usetex=False)
rc('font', family='serif', size=34)
rc('figure', figsize=(16,16))
rc('axes',linewidth=2)
```

```python
df = pd.read_csv("data/clusters.csv")
layer = 5
df_layer = df[df["layer"]==layer]
x = df_layer["x"]
y = df_layer["y"]

plt.scatter(x, y, s=4)
plt.xlabel("x [cm]", fontsize=38)
plt.ylabel("y [cm]", fontsize=38)
plt.savefig("whole_detector.png", dpi=300)
plt.show
```

```python
df = pd.read_csv("data/clusters.csv")
layer = 5
df_layer = df[df["layer"]==layer]

clusters  = df_layer[df_layer["clusterId"] != -1]
outliers = df_layer[df_layer["clusterId"] == -1]

clusters = clusters.groupby(["clusterId"])
for _, cluster in clusters:
    plt.plot(cluster.x, cluster.y, marker="o", markersize=4, linestyle="")
plt.plot(outliers.x, outliers.y, c='black', marker='s', markersize=1, linestyle="")


plt.xlabel("x [cm]", fontsize=38)
plt.ylabel("y [cm]", fontsize=38)
plt.savefig("clusters_whole_detector.png",)
plt.show()
```

```python
df = pd.read_csv("data/clusters.csv")
layer = 5
df_layer = df[df["layer"]==layer]

energies = df_layer['weight'].to_numpy()
min_e = np.min(energies)
max_e = np.max(energies)
min_col = 0
max_col = 5
colors = []
for i in range(energies.size):
    color = ((energies[i] - min_e) / (max_e - min_e)) * (max_col - min_col) + min_col
    colors.append(color)

#print(colors)

plt.scatter(df_layer['x'], df_layer['y'], c=colors, cmap='magma', s=2)
plt.colorbar()


plt.show()
```

```python
df = pd.read_csv("data/clusters.csv")
layer = 5
df_layer = df[df["layer"]==layer]

energies = df_layer['weight'].to_numpy()
min_e = np.min(energies)
max_e = np.max(energies)

colormap = plt.cm.tab20c
normalize = mpl.colors.PowerNorm(gamma=0.3)
#normalize = mpl.colors.LogNorm(vmin=min_e, vmax=max_e)

plt.scatter(df_layer['x'], df_layer['y'], c=energies, cmap=colormap, norm=normalize, s=2)
plt.colorbar(shrink=1, ticks=(0,0.01,0.1,0.25,0.5,1,3,5), extend='max', spacing='uniform', label='Energy of the hit [GeV]')
plt.xlabel("x [cm]")
plt.ylabel("y [cm]")

plt.show()
```
