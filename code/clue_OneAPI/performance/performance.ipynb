```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc
from matplotlib import gridspec

rc('text', usetex=False)
rc('font', family='serif', size=18)
rc('figure', figsize=(12,8))
rc('axes',linewidth=2)
PointsPerLayer = np.linspace(1,10,10)
```

```python
time_cpu = pd.read_csv("cpu.csv")["mean"]
err_cpu = pd.read_csv("cpu.csv")["std"]

time_tbb = pd.read_csv("TBB.csv")["mean"]
err_tbb = pd.read_csv("TBB.csv")["std"]

time_gpu = pd.read_csv("gpu_CUDA.csv")["mean"]
err_gpu = pd.read_csv("gpu_CUDA.csv")["std"]

time_sycl_cuda = pd.read_csv("gpu_SYCL_CUDA.csv")["mean"]
err_sycl_cuda = pd.read_csv("gpu_SYCL_CUDA.csv")["std"]

time_sycl_cpu = pd.read_csv("SYCL_Intel_cpu.csv")["mean"]
err_sycl_cpu = pd.read_csv("SYCL_Intel_cpu.csv")["std"]

time_sycl_gpu = pd.read_csv("SYCL_Intel_gpu.csv")["mean"]
err_sycl_gpu = pd.read_csv("SYCL_Intel_gpu.csv")["std"]
```

```python
fig,ax = plt.subplots()
ax.errorbar(PointsPerLayer, time_cpu, err_cpu, label="CPU", 
            fmt = 'o-', color='#FF6A13', mfc='#FF6A13', mec="#FF6A13", ecolor='#FF6A13', alpha=0.6, capsize=5) 
ax.set_xlabel("Number of points per layer ($\cdot 10^3$)", fontsize = 16)
ax.set_ylabel("Total CPU execution time [ms]", fontsize=16)
ax.set_yscale("log")

ax2=ax.twiny()
ax2.axes.xaxis.set_visible(False)
ax2.errorbar(PointsPerLayer, time_sycl_cpu, err_sycl_cpu, label="SYCL CPU",
            fmt = 'o-', color='dodgerblue', mfc='dodgerblue', mec="dodgerblue", ecolor='dodgerblue', capsize=5)

ax3=ax.twiny()
ax3.axes.xaxis.set_visible(False)
ax3.errorbar(PointsPerLayer, time_tbb, err_tbb, label="TBB[48T]",
            fmt = 'o-', color='red', mfc='red', mec="red", ecolor='red', capsize=5)

fig.suptitle("Intel Xeon Gold 6336Y", fontsize=18)
fig.legend(loc=(0.10,0.76), fontsize=14)
plt.show()
```
